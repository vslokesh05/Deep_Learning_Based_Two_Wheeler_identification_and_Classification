```python
def get_info(image_path):
  text_path = image_path.replace('images', 'labels').replace('.jpg', '.txt')
  annots = pd.read_csv(text_path, sep=' ', header=None).values
  classes = list(map(int, annots[:, 0]))
  annots = annots[:, 1:]
  return image_path.split('/')[-1], classes, annots
```

```python
def yolo_to_tf(annot, size=1280):
  x_c, y_c, width, height = list(map(int, annot * size))
  x_min = x_c - (width // 2)
  x_max = x_c + (width // 2)
  y_min = y_c - (height // 2)
  y_max = y_c + (height // 2)
  return np.array([x_min, x_max, y_min, y_max]) / size
```

```python
from object_detection.utils import dataset_util
from object_detection.dataset_tools import tf_record_creation_util
import cv2
import math

def create_tf_example(image_path):
    filename, classes_, annotations = get_info(image_path)

    with tf.io.gfile.GFile(image_path, 'rb') as image_bin:
        image_bin = image_bin.read()
    encoded_jpg_io = io.BytesIO(image_bin)

    xmins = []
    xmaxs = []
    ymins = []
    ymaxs = []
    classes_text = []
    classes = []
    image_format = 'jpg'.encode('utf8')

    # temp_im = cv2.imread(image_path)
    # height, width, channels = temp_im.shape
    height, width = 640, 640

    for ij in range(len(annotations)):
        xmin, xmax, ymin, ymax = yolo_to_tf(annotations[ij], size=640)
        xmins.append(xmin)
        xmaxs.append(xmax)
        ymins.append(ymin)
        ymaxs.append(ymax)
        classes_text.append(str(classes_[ij] + 1).encode('utf8'))
        classes.append(classes_[ij])

    filename = filename.encode('utf8')

    tf_example = tf.train.Example(features=tf.train.Features(feature={
      'image/height': dataset_util.int64_feature(height),
      'image/width': dataset_util.int64_feature(width),
      'image/filename': dataset_util.bytes_feature(filename),
      'image/source_id': dataset_util.bytes_feature(filename),
      'image/format': dataset_util.bytes_feature(image_format),
      'image/encoded': dataset_util.bytes_feature(image_bin),
      'image/object/bbox/xmin': dataset_util.float_list_feature(xmins),
      'image/object/bbox/xmax': dataset_util.float_list_feature(xmaxs),
      'image/object/bbox/ymin': dataset_util.float_list_feature(ymins),
      'image/object/bbox/ymax': dataset_util.float_list_feature(ymaxs),
      'image/object/class/text': dataset_util.bytes_list_feature(classes_text),
      'image/object/class/label': dataset_util.int64_list_feature(classes),
    }))

    return tf_example
```

```python
from tqdm import tqdm
def convert_to_tfrecord(files, file_name, num_shards=10):
    with contextlib2.ExitStack() as tf_record_close_stack:
        output_tfrecords = tf_record_creation_util.open_sharded_output_tfrecords(tf_record_close_stack, file_name, num_shards)
        for index, image_path in enumerate(tqdm(files)):
          try:
            tf_example = create_tf_example(image_path)
          except:
            continue
          output_shard_index = index % num_shards
          output_tfrecords[output_shard_index].write(tf_example.SerializeToString())
```

```python
# Define some utils method for prediction.

def load_image_into_numpy_array(path):
    img_data = tf.io.gfile.GFile(path, 'rb').read()
    image = Image.open(io.BytesIO(img_data))
    (im_width, im_height) = image.size
    return np.array(image.getdata()).reshape(
      (im_height, im_width, 3)).astype(np.uint8)

def detect(image_np):
    """Detect COTS from a given numpy image."""
    input_tensor = image_np[np.newaxis, ...]
    # start_time = time.time()
    detections = model(input_tensor)
    return detections
```

```python
DETECTION_THRESHOLD = 0.1
height, width = 640, 640
num_detections = detections['num_detections'][0].numpy().astype(np.int32)
predictions = []
for index in range(num_detections):
    score = detections['detection_scores'][0][index].numpy()
    if score < DETECTION_THRESHOLD:
        continue

    bbox = detections['detection_boxes'][0][index].numpy()
    y_min = int(bbox[0] * height)
    x_min = int(bbox[1] * width)
    y_max = int(bbox[2] * height)
    x_max = int(bbox[3] * width)

    bbox_width = x_max - x_min
    bbox_height = y_max - y_min

    predictions.append('{:.2f} {} {} {} {}'.format(score, x_min, y_min, bbox_width, bbox_height))

# Generate the submission data.
prediction_str = ' '.join(predictions)
# sample_prediction_df['annotations'] = prediction_str

print('Prediction:', prediction_str)
```
